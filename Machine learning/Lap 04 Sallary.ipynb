## IMPORT LIBRARY

```python
import numpy as np # membaca angka
import pandas as pd # memproses membaca data
import matplotlib.pyplot as plt # memvisualisasi data 
```

## MEMBACA DATA SET

```python
data = pd.read_csv('salary_dataset.csv')
data 
```

Di sini saya menggunakan data set sallary

```python
data.info()
```

data info untuk mengetahui info dari data set

```python
data.corr()
```

```python
data = data.drop('no', axis=1)
data
```

disini saya men drop data no karena tidak di perlukan 

```python
x_train = data['YearsExperience'].values[:,np.newaxis]
y_train = data['Salary'].values
```

```python
plt.scatter(data.YearsExperience, data.Salary)
plt.xlabel('Lama Pengalaman')
plt.ylabel('Gaji')
plt.title('Plot Data Years Experience dan Salary')
plt.show()
```

```python
YearsExperience_baru = ([1.1],[11],[7])
```

Di sini saya memasukkan angka sesuai dengan perintah 

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train, y_train)

prediksi_baru =model.predict(YearsExperience_baru)
print(prediksi_baru)
```

## VISUALISASI DATA

```python
pb = model.predict(x_train)
plt.scatter(data.YearsExperience, data.Salary)
plt.plot(x_train, pb, color = "red", linewidth = 1)
plt.xlabel('Lama Pengalaman')
plt.ylabel('Gaji')
plt.title('Plot Data Years Experience dan Salary')
plt.show()
```
